# Databricks

## 1. Create catalog, volume and schema

### SparkSQL

```python
%sql

-- 1. See what catalogs and what schemas exists 
SHOW CATALOGS;
SHOW SCHEMAS IN catalog_name;


-- 2. Create CATALOG
-- best practice är att skapa en catalog per projekt - 'data' är första catalogen vi skapade i denna kurs
CREATE CATALOG catalog_name;

-- 3.Create SCHEMA in your catalog
CREATE SCHEMA IF NOT EXISTS catalog_name.schema_name
COMMENT 'optional description';

-- 4. Create VOLUME under that schema
-- fully qualified name: catalog.schema.volume
CREATE VOLUME IF NOT EXISTS catalog_name.schema_name.raw_files;

-- 5. List volume -> see what data exists
LIST '/Volumes/catalog_name/schema_name/volume_name';
```

## 2. Upload data into a VOLUME

**Using the Catalog Explorer UI**
Navigate to Catalog Explorer (left sidebar)
Browse to: data → nobel → raw_files (your volume)
Click "Upload to Volume" button in the UI
Select your CSV files from the Kaggle dataset
Files will be uploaded to /Volumes/data/nobel/raw_files/
