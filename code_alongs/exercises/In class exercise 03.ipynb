## 1. Create catalog, volume and schema

### Create CATALOG 
NA since we are using the 'data' catalog for this exercise

### Create SCHEMA

````python
%sql
-- 1.Create schema in your catalog
CREATE SCHEMA [IF NOT EXISTS] catalog_name.schema_name
COMMENT 'optional description';
````

````python
%sql
CREATE SCHEMA if not EXISTS data.nobel COMMENT 'Schema for Nobel Prize analysis';
````

### Create VOLUME

````python
%sql
-- 2. Create volume under that schema
CREATE VOLUME IF NOT EXISTS catalog_name.schema_name.raw_files;
````

````python
%sql
CREATE VOLUME IF NOT EXISTS data.nobel.raw_files;
````

````python
%sql
SHOW SCHEMAS IN data;
````

## 2. Upload & read in data
1. Uploaded through Catalog Explorer UI
2. Read in the CSV in to Spark

````python
%sql
-- 1. After uploading the files
-- See what files we have in our catalog
LIST '/Volumes/data/nobel/raw_files';
````

````python
# 2. Read in CSV
DATA_PATH = '/Volumes/data/nobel/raw_files'

df_nobel = spark.read.csv(f"{DATA_PATH}/complete.csv", header=True, inferSchema=True)
df_nobel

````
